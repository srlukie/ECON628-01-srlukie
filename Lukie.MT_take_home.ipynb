```python
iowa['Date']=pd.to_datetime(iowa['Date'])
iowa = iowa.set_index(pd.DatetimeIndex(iowa['Date']))
```

```python
iowa.head()
```

```python
iowa.dropna(subset = ['County Number', 'County', 'Category', 'Category Name'], inplace = True)
#print iowa.isnull().sum()
```

```python
#Object to Float: http://widequestion.com/question/pandas-column-convert-currency-to-float/ 
iowa['Sale (Dollars)'] = (iowa['Sale (Dollars)'].replace( '[\$,)]','', regex=True ).astype(float))
iowa['State Bottle Cost'] = (iowa['State Bottle Cost'].replace( '[\$,)]','', regex=True ).astype(float))
iowa['State Bottle Retail'] = (iowa['State Bottle Retail'].replace( '[\$,)]','', regex=True ).astype(float))

#Float to Int 
iowa['County Number'] = iowa['County Number'].astype(int)
iowa['Category'] = (iowa['Category']).astype(int)

#iowa.info()
```

```python
st_15 = iowa15.groupby('Store Number')['Month'].agg(['max','min'])
st_15['diff'] = st_15['max']-st_15['min'] + 1
st_15['diff']
```

```python
iowa15 = iowa15.set_index('Store Number')
```

```python
iowa15['filter'] = (st_15['diff'] > 11).astype(int)
```
